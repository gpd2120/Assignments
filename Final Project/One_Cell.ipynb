```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import svd
from scipy.linalg import inv

# Load Station Data 
station_data = np.genfromtxt('Stations_Seismology.txt', dtype=None, encoding=None)
station_names = np.array([row[0] for row in station_data])  
latitude = np.array([float(row[1]) for row in station_data]) 
longitude = np.array([float(row[2]) for row in station_data])  
elevation = np.array([float(row[3]) for row in station_data]) 

# Load Phase Data
phase_data = np.genfromtxt('Phases_Seismology.txt', dtype=None, encoding=None)
stationpick = np.array([row[0] for row in phase_data])  
arrival = np.array([float(row[1]) for row in phase_data])  
quality = np.array([float(row[2]) for row in phase_data])  
phase = np.array([row[3] for row in phase_data])  

# Load CA Fault Lines 
CAfaults_data = np.loadtxt('CAfaults.dat.WGCEP')
CA_data = np.loadtxt('california.lin')

# Constants for Calculations
origin_lat = 37.852 
origin_lon = -122.215 
p_velocity = 5.5  
lat2km = 111.1  
lon2km = 88.8  
sigma = np.var(arrival) 

nph = len(phase)
matched_indices = np.isin(station_names, stationpick)  
lo = longitude[matched_indices]
la = latitude[matched_indices]
el = elevation[matched_indices]

# Initial Guesses for time, depth, lat, lon
toO = 2  
deO = 5  
laO = origin_lat  
loO = origin_lon 

runs = 5  # number of iterations

# Initialize 
dframe = np.zeros((4, runs + 1))  
dframe[:, 0] = [toO, loO, laO, deO]

# Plotting Faults
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(CAfaults_data[:, 1], CAfaults_data[:, 0], 'darkblue', label="CA Faults")
ax.plot(CA_data[:, 1], CA_data[:, 0], 'black', label="California Topography")

# Plotting Stations
ax.set_aspect(0.88 / 1.11)
ax.plot(longitude, latitude, '.', markeredgecolor='royalblue', markerfacecolor='lightskyblue', markersize=10)
for i, name in enumerate(station_names):
    ax.text(longitude[i] + 0.02, latitude[i], str(name))

# Plotting Initial Guess
ax.set_xlim([-122.6, -121.7])
ax.set_ylim([37.55, 38.2])
ax.plot(origin_lon, origin_lat, 'o', markeredgecolor='rebeccapurple', markerfacecolor='plum', markersize=11, label="Initial Location")

# Depth data for plotting
depths = [deO]
residuals_all = []

# Iteration and change to cartesian coordinates 
for u in range(runs):
    x = lon2km * (lo - loO)  
    y = lat2km * (la - laO)  
    z = el - deO  

# Distance
    d0 = np.sqrt(x**2 + y**2 + z**2)
    r = arrival - toO - (d0 / p_velocity) 
    residuals_all.extend(r)  
    
# Matrix
    G = np.zeros((nph, 4))
    G[:, 0] = 1
    G[:, 1] = x / (d0 * p_velocity)  
    G[:, 2] = y / (d0 * p_velocity)
    G[:, 3] = z / (d0 * p_velocity)

# SVD
    U, S, Vt = svd(G, full_matrices=False)
    dm = np.linalg.pinv(G) @ r
    covmatrix = np.diag(sigma * np.linalg.inv(G.T @ G))

# Errors for each parameter
    err_toO = np.sqrt(covmatrix[0]) 
    err_loO = np.sqrt(covmatrix[1]) / lon2km
    err_laO = np.sqrt(covmatrix[2]) / lat2km
    err_deO = np.sqrt(covmatrix[3])  

# Updating parameters
    toO -= dm[0]
    loO -= (dm[1] / lon2km)
    laO -= (dm[2] / lat2km)
    deO -= dm[3]

# Store depth for plotting
    depths.append(deO)

# Storing parameters
    dframe[:, u + 1] = [toO, loO, laO, deO]

# Plotting change in location
    ax.plot(loO, laO, 'h', markeredgecolor='rebeccapurple', markerfacecolor='slateblue', markersize=10)
    ax.errorbar(loO, laO, yerr=err_laO, color='black', fmt='none')
    ax.errorbar(loO, laO, xerr=err_loO, color='black', fmt='none')

# Print 
    print(f"Iteration {u + 1}: toO = {toO}, loO = {loO}, laO = {laO}, deO = {deO}")

# Plot this iteration's location on the map with distinct markers
    ax.plot(loO, laO, 'o', markeredgecolor='black', markerfacecolor='mediumorchid', markersize=5, label=f"Iteration {u + 1}")

# Final location
ax.plot(loO, laO, '*', markeredgecolor='darkmagenta', markerfacecolor='fuchsia', markersize=13, label="Final Location")
ax.legend()
plt.title("Earthquake Location")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# Plot Depth 
plt.figure(figsize=(12, 8))
plt.plot(range(runs + 1), depths, '-o', color='palevioletred', markerfacecolor='lightpink')
plt.xlabel("Iteration")
plt.ylabel("Depth (km)")
plt.title("Depth Section")
plt.grid()
plt.show()

# Plot Residual Distribution
plt.figure(figsize=(12, 8))
plt.hist(residuals_all, bins=20, color='lavender', edgecolor='mediumslateblue')
plt.xlabel("Residual (s)")
plt.ylabel("Frequency")
plt.title("Residual Distribution")
plt.grid()
plt.show()
```
