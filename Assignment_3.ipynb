## 1. Creating and Manipulating Arrays

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

#### 1.1

```python
x = np.linspace(-2,2, 100)
y = np.linspace(-2,2, 100)
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
x,y
```

#### 1.2

```python
plt.pcolormesh(xx)
```

```python
plt.pcolormesh(yy)
```

#### 1.3

```python
radial=np.sqrt(xx**2 + yy**2)
angle=np.arctan2(yy,xx)
radial, angle
```

#### 1.4

```python
plt.pcolormesh(xx,yy, radial)
```

```python
plt.pcolormesh(xx,yy, angle)
```

#### 1.5

```python
f = (np.cos(4*radial))**2 + (np.sin(4*angle))**2
```

```python
plt.pcolormesh(xx,yy, f)
```

#### 1.6

```python
f.mean()
```

```python
f_xmean = f.mean(axis=1)
plt.plot(f_xmean, y);
y
f_xmean
```

```python
f_ymean = f.mean(axis=0)
plt.plot(x, f_ymean);
f_ymean
x
```

```python
meanS = S.mean (axis=1)
meanS
```

#### 2.5

```python
plt.plot(meanS, levels)
plt.errorbar(meanS, levels, xerr = stdS)
plt.gca().invert_yaxis()
plt.grid (True, c = 'black')
```
